```python
import glob
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
import xml.etree.ElementTree as ET
from random import randint
import random
```

```python
def file_lines_to_list(path):
    # open txt file lines to a list
    with open(path) as f:
        content = f.readlines()
    # remove whitespace characters like `\n` at the end of each line
    content = [(x.strip()).split() for x in content]
    return content

def get_file_name(path):
    basename = os.path.basename(path)
    onlyname = os.path.splitext(basename)[0]
    return onlyname
```

```python
def display_rgb(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# Test Drawing Boxes

```python
enhanced_ver2_6080anchors = glob.glob("./graphs/60_70_enhanced_ver2_6080anchors_yolo_epoch_92/*.txt")
```

```python
tested_by_nonsmall_20 = glob.glob("./graphs/60_70_tested_by_nonsmall_20/*.txt")
```

```python
index = randint(0, len(enhanced_ver2_6080anchors)-1)
name = get_file_name(enhanced_ver2_6080anchors[index])
print(name)

img = plt.imread(f"../datasets/bdd100k/bdd100k/bdd100k/images/100k/val/{name}.jpg")

gt_boxes = file_lines_to_list(f"./yolo-various-size/60-70/bdd100k_60_70_GT/{name}.txt")
det_boxes = file_lines_to_list(f"./graphs/60_70_enhanced_ver2_6080anchors_yolo_epoch_92/{name}.txt")

print(len(gt_boxes), len(det_boxes))

dummy = img.copy()

for obj in gt_boxes:
    dummy = cv2.rectangle(dummy, (int(obj[1]), int(obj[2])), (int(obj[3]), int(obj[4])), (255, 0, 0), thickness = 2)
    
for obj in det_boxes:
    dummy = cv2.rectangle(dummy, (int(obj[2]), int(obj[3])), (int(obj[4]), int(obj[5])), (0, 255, 0), thickness = 2)
                                                               
plt.figure(figsize = (15, 10))
plt.imshow(dummy)
plt.show()
```

```python
det_boxes_20 = file_lines_to_list(f"./graphs/60_70_tested_by_nonsmall_20/{name}.txt")

print(len(gt_boxes), len(det_boxes_20))

dummy_20 = img.copy()

for obj in gt_boxes:
    dummy_20 = cv2.rectangle(dummy_20, (int(obj[1]), int(obj[2])), (int(obj[3]), int(obj[4])), (255, 0, 0), thickness = 2)
    
for obj in det_boxes_20:
    dummy_20 = cv2.rectangle(dummy_20, (int(obj[2]), int(obj[3])), (int(obj[4]), int(obj[5])), (0, 255, 0), thickness = 2)
                                                               
plt.figure(figsize = (15, 10))
plt.imshow(dummy_20)
plt.show()
```
