```python
boxes = [[100, 100, 250, 300], [800, 300, 1000, 500]]
```

```python
one_mask = np.ones([720, 1280, 3])

# plt.figure(figsize=(15,10))
plt.imshow(one_mask)
plt.show()
```

```python
# one_mask = one_mask*255.0
mask = one_mask.copy()

for box in boxes:
    cv2.rectangle(mask, (box[0], box[1]), (box[2], box[3]), (0, 0, 0), -1)

plt.imshow(mask)
plt.show()
```
